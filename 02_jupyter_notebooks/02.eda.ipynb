```python
# Data distribution of numerical features (histograms with trend lines) - code created using Chat GPT 
# Selecting numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Creating subplots with improved layout and styling
fig = sp.make_subplots(
    rows=len(numerical_columns), cols=1, 
    subplot_titles=[f'Distribution of {col}' for col in numerical_columns],
    vertical_spacing=0.08
)

# Colorblind-friendly palette
colorblind_palette = ["#377eb8", "#ff7f00", "#4daf4a", "#f781bf", "#a65628", "#984ea3", "#999999"]

# Adding histograms and trend lines
for i, column in enumerate(numerical_columns):
    hist_data = df[column].dropna()
    color = colorblind_palette[i % len(colorblind_palette)]  # Cycle through the palette
    
    # Histogram trace
    fig.add_trace(
        go.Histogram(
            x=hist_data, 
            nbinsx=30,  # Increased bins for better granularity
            name=column, 
            opacity=0.85,  # Slightly increased opacity for better visibility
            marker=dict(color=color, line=dict(color='black', width=1.5))  # High contrast borders
        ),
        row=i+1, col=1
    )
    
    # KDE trend line
    kde = gaussian_kde(hist_data)
    x_grid = np.linspace(hist_data.min(), hist_data.max(), 1000)  # Grid for the KDE curve
    kde_values = kde(x_grid)
    
    # Trend line trace (KDE curve)
    fig.add_trace(
        go.Scatter(
            x=x_grid, 
            y=kde_values * (hist_data.count() * np.diff(np.histogram(hist_data, bins=30)[1])[0]),  # Scale KDE to match histogram
            mode='lines',
            name=f'{column} Trend Line',
            line=dict(color=color, width=2, dash='dash')  # Dashed line for trend
        ),
        row=i+1, col=1
    )

# Updating layout for better visualization
fig.update_layout(
    height=350 * len(numerical_columns),  # Increased height for better spacing
    width=1000,  # Slightly wider for better readability
    showlegend=False, 
    title_text="Distributions of Numerical Features with Trend Lines", 
    title_font=dict(size=22, family="Arial", color="black"),
    template="plotly",  # Set to a more colorblind-friendly template
    paper_bgcolor="#F5F5F5",  # Light grey background for reduced glare
    plot_bgcolor="#EAEAEA",  # Slightly darker grey for plot area
    margin=dict(l=60, r=60, t=90, b=60),
    bargap=0.2  # Added gap between bars for better differentiation
)

fig.show()
```
